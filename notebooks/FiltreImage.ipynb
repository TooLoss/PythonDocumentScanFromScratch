```python

from src import main as m, filter as f
import numpy as np
```

```python
def Bruit(n):
    R = np.full((n,n), 255)
    for i in range(n):
        for j in range(n):
            R[i,j] = R[i,j] - (i+j)%7
    return R
```

```python
B = Bruit(50)
m.AfficherRange(B)
```

```python
S = f.Sauvola(B, 5, .2)
m.AfficherRange(S)
```
