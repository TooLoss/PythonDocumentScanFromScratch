```python

import matplotlib.pyplot as plt
from src import ocr as o
from src.data import data
```

```python
k_test = [k for k in range(1, 10)]
DTrace = {k:0 for k in k_test} # Dictionnaire trace
```

```python
for k in k_test:
    DTrace[k] = o.Confusion(data, k)
```

```python
Traces = list(DTrace.items())
Traces = sorted(Traces, key=lambda x: x[1], reverse=True)
```

```python
print("Meilleur valeur de k : " + str(Traces[0][0]))
```

```python
X, Y = zip(*Traces)
```

```python
plt.scatter(X, Y, color='red')
plt.title("Trace de la matrice confusion en fonction de k")
plt.xlabel("Choix de k")
plt.ylabel("Trace de la matrice confusion")
plt.show()
```
